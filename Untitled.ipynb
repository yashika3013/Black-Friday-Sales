```python
data = df.copy
```

```python
data = pd.get_dummies(df, columns=['Stay_In_Current_City_Years'])
```

## Encoding the categorical variables

```python
from sklearn.preprocessing import LabelEncoder
lr = LabelEncoder()
```

```python
data['Gender'] = lr.fit_transform(data['Gender'])
```

```python
data['Age'] = lr.fit_transform(data['Age'])
```

```python
data['City_Category'] = lr.fit_transform(data['City_Category'])
```

```python
data.head()
```

```python
data['Product_Category_2'] =data['Product_Category_2'].fillna(0).astype('int64')
data['Product_Category_3'] =data['Product_Category_3'].fillna(0).astype('int64')
```

```python
df.isnull().sum()
```

```python
df.info()
```

## Dropping irrelevant cols

## Splitting data into independent and dependent variables

```python
X = data.drop(['Purchase'], axis =1)
y = data['Purchase']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state = 123)
```

## Modelling 
### Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
from math import sqrt
print("RMSE of Linear Regression Model is ",sqrt(mean_squared_error(y_test, y_pred)))
```
